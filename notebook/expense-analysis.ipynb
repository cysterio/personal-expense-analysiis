```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

path = "/Users/Prasakha/personal-expense-analysis/data/expenses.csv"

df = pd.read_csv(path)
df

# df.groupby("category")["amount"].sum()
df["date"] = pd.to_datetime(df["date"])
daily_spending = df.groupby("date")["amount"].sum()
daily_spending



```

```python
daily_spending.index = daily_spending.index.astype(str)
daily_spending.plot(kind="bar", title="Daily Spending")
plt.ylabel("Amount")
plt.show()
```

```python
category_spending = df.groupby("category")["amount"].sum()
category_spending
category_spending.plot(
    kind="pie",
    autopct="%1.1f%%",
    title="Spending Distribution by Category"
)
plt.ylabel("")
plt.show()
```
